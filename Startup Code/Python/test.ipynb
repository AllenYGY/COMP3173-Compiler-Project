```python
with open('../../SLR Parsing Table.csv', 'r') as file:
    lines = file.readlines()

# 去掉每行的尾部换行符，并以逗号分割
data = [line.strip().split(',') for line in lines]

# 第一行可能是标题行，可以忽略。假设第二行是列名，第三行开始是数据
header = data[1][1:]  # 从第二行提取列名，忽略状态编号列
print("Header:", header)
# 忽略第一列（状态编号），从第三行开始读取数据
data = [row[1:] for row in data[2:]]

# 查找列名 'S'' 的位置以分割ACTION和GOTO
idx = header.index("S'")

# 分割出ACTION部分的列名
terminals = header[:idx]
# print("Terminals:", terminals)

# 分割ACTION表，需要遍历每一行，取出前idx列
ACTION_TABLE = [row[:idx] for row in data]
GOTO_TABLE = [row[idx:] for row in data]



ACTION = {} 
# 获取动作和符号的列名
# 遍历每行数据来填充ACTION字典
for state, row in enumerate(ACTION_TABLE):
    for col_index, action in enumerate(row):
        if action:  # 如果单元格内有动作信息，则处理
            # 解析动作和值，动作可以是's', 'r', 或 'acc'
            act_type = action[0]  # 取第一个字符为动作类型
            if act_type in {'s', 'r'}:
                # 对于's'或'r', 后续字符为状态编号或产生式编号
                act_value = int(action[1:])  # 取剩余部分转换为整数
                ACTION[(state, terminals[col_index])] = (act_type, act_value)
            elif action == 'acc':
                # 如果是接受动作，则不需要值
                ACTION[(state, terminals[col_index])] = (action, None)


# 初始化GOTO字典
GOTO = {}

# 获取非终结符的列名
non_terminals = header[idx:]
print("Non-terminals:", non_terminals)

# 遍历每行数据来填充GOTO字典
for state, row in enumerate(data):
    for col_index, next_state in enumerate(row[idx:]):
        if next_state.strip():  # 检查单元是否非空
            # next_state应该是一个整数，表示转移后的状态
            GOTO[(state, non_terminals[col_index])] = int(next_state)

# print("GOTO:", GOTO)

# 打印GOTO字典查看结果
for key, value in GOTO.items():
    print(f"{key}: {value}")


```

```python
def read_csv(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    data = [line.strip().split(',') for line in lines]
    return data

def parse_header_and_data(data):
    header = data[1][1:]  # 第二行是列名，忽略状态编号列
    data = [row[1:] for row in data[2:]]  # 忽略第一列（状态编号），从第三行开始读取数据
    return header, data

def split_action_goto(data, header):
    idx = header.index("S'")
    terminals = header[:idx]
    non_terminals = header[idx:]
    ACTION_TABLE = [row[:idx] for row in data]
    GOTO_TABLE = [row[idx:] for row in data]
    return terminals, non_terminals, ACTION_TABLE, GOTO_TABLE

def parse_action(ACTION_TABLE, terminals):
    ACTION = {}
    for state, row in enumerate(ACTION_TABLE):
        for col_index, action in enumerate(row):
            if action:  # 如果单元格内有动作信息，则处理
                act_type = action[0]  # 动作类型
                if act_type in {'s', 'r'}:
                    act_value = int(action[1:])  # 状态编号或产生式编号
                    ACTION[(state, terminals[col_index])] = (act_type, act_value)
                elif action == 'acc':
                    ACTION[(state, terminals[col_index])] = (action, None)
    return ACTION

def parse_goto(GOTO_TABLE, non_terminals):
    GOTO = {}
    for state, row in enumerate(GOTO_TABLE):
        for col_index, next_state in enumerate(row):
            if next_state.strip():  # 检查单元是否非空
                GOTO[(state, non_terminals[col_index])] = int(next_state)
    return GOTO

def print_table(table):
    for key, value in table.items():
        print(f"{key}: {value}")

def read_grammar(file_path: str) -> dict[int, dict]:
    grammar = {}
    with open(file_path, 'r') as file:
        lines = file.readlines()

    for line in lines:
        line = line.strip()
        if line:
            parts = line.split('.')
            if len(parts) > 1:
                rule_part = parts[1].strip()
                lhs_rhs = rule_part.split('->')
                if len(lhs_rhs) == 2:
                    lhs = lhs_rhs[0].strip()  
                    rhs = lhs_rhs[1].strip() 
                    rule_number = int(parts[0].strip())  
                    rhs_tmp = rhs.split()
                    length = len(rhs_tmp)  
                    # grammar[rule_number] = {'non-terminal': lhs, 'productions': rhs, 'length': length}
                    grammar[rule_number] = {'non-terminal': lhs, 'length': length}

    return grammar

if __name__ == '__main__':
    file_path = '../../SLR Parsing Table.csv'
    data = read_csv(file_path)
    header, data = parse_header_and_data(data)
    terminals, non_terminals, ACTION_TABLE, GOTO_TABLE = split_action_goto(data, header)
    print("Terminals:", terminals)
    print("Non-terminals:", non_terminals)
    ACTION = parse_action(ACTION_TABLE, terminals)
    GOTO = parse_goto(GOTO_TABLE, non_terminals)
    # print("ACTION Table:")
    # print_table(ACTION)
    print("GOTO Table:")
    print_table(GOTO)
```
