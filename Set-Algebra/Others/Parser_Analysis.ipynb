```python
import pandas as pd

# ignore the first row and column of the csv file
df = pd.read_csv('../../SLR Parsing Table.csv', index_col=0,skiprows=1)

df = df.fillna('')
# find the column ['$'] and split the table  into two parts
idx = df.columns.get_loc("S'")
ACTION = df.iloc[:, :idx]
GOTO = df.iloc[:, idx:]
# convert the value in goto table except the empty string to integer
for column in GOTO.columns:
    GOTO[column] = GOTO[column].apply(lambda x: int(x) if x != '' else x)


```

```python
# 读取CSV文件
with open('../../SLR Parsing Table.csv', 'r') as file:
    lines = file.readlines()

# 去掉每行的尾部换行符，并以逗号分割
data = [line.strip().split(',') for line in lines]

# 忽略第一行和第一列
data = [row[1:] for row in data[1:]]

# 查找列名 'S'' 的位置
idx = data[0].index("S'")

# 将数据分为 ACTION 和 GOTO 两部分
ACTION = [row[:idx] for row in data]
GOTO = [row[idx:] for row in data]

# # GOTO 部分需要将非空字符串转换为整数
for i in range(len(GOTO)):
    if i == 0:
        continue
    for j in range(len(GOTO[i])):
        if GOTO[i][j].strip():
            GOTO[i][j] = int(GOTO[i][j])

# 打印前5行以检查
print("ACTION:")
for row in ACTION[:5]:
    print(row)

print("\nGOTO:")
for row in GOTO[:5]:
    print(row)
```

```python
def read_and_parse_grammar(file_path):
    print(file_path)
    grammar = {}
    # Open and read the grammar file
    with open(file_path, 'r') as file:
        lines = file.readlines()
    # Iterate through each line to parse grammar rules
    for line in lines:
        line = line.strip()
        if line:  # If line is not empty
            parts = line.split('.')
            # convert the first part to integer
            rule_part = parts[1].strip()  # Get "S' -> S" part
            lhs_rhs = rule_part.split('->')
            lhs = lhs_rhs[0].strip()  # Left-hand side (non-terminal)
            rhs = lhs_rhs[1].strip()  # Right-hand side (production)
            rule_number = int(parts[0].strip())  # Rule number (converted to integer)
            rhs_tmp=rhs.split()
            length=len(rhs_tmp)
            grammar[rule_number] = {'non-terminal': lhs, 'productions': rhs, 'length': length}
    return grammar
# Assume the file path is correct
grammar_file_path = '../../SLR Grammar.txt'

grammar_rules = read_and_parse_grammar(grammar_file_path)

# Print parsed grammar for verification
for index, details in grammar_rules.items():
    print(f"Rule {index}: {details['non-terminal']} -> {details['productions'],details['length']}")
```
